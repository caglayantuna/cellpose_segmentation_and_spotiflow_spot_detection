```python
import matplotlib.pyplot as plt
from skimage.io import imread
from spotiflow.model import Spotiflow
import bigfish.plot as plot
```

```python
model_spot = Spotiflow.from_pretrained("general")
```

```python
image = imread("./data/sample_image_c2.png")
```

```python
points, details = model_spot.predict(image)
```

# Image and detected spots

```python
plot.plot_detection(image, points, contrast=True)
```
